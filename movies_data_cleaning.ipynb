```python
import pandas as pd
```

```python
#Read data from file
df = pd.read_csv(r'I:\Python Learnings\Bechdel Data\movies.csv')
df.head(2)
```

```python
#drop columns
df = df.drop(columns=['id','keywords','runtime','popularity','cast','homepage','tagline','overview','production_companies','budget_adj','revenue_adj'])
df.head()
```

```python
df.dtypes
```

```python
df['director'] = df['director'].str.strip()
df['release_date'] = pd.to_datetime(df['release_date'])
df.dtypes
```

```python
df.head(2)
```

```python
df['genres'] = df['genres'].str.split('|')
df.head(2)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(how = 'any', subset=['imdb_id','director','genres'])
df.isnull().sum()
```

```python
int(df['imdb_id'].duplicated().sum())
```

```python
df = df.drop_duplicates(subset='imdb_id', keep='first')
int(df['imdb_id'].duplicated().sum())
```

```python
df.head()
```

```python
df.to_csv(r'I:\Python Learnings\Bechdel Data\Cleaned_movies.csv', index=False)
```
